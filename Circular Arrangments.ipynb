```
import math

# Algorithmic Implementation of Burnside's Lemma

# Finds common factors of a list of numbers
def FindDivisors(aList):
  hcf = aList[0]
  Divisors = []

  # Find highest common factor
  for i in range(1, len(aList)): 
    hcf = math.gcd(aList[i], hcf) 
  
  # Find factors
  for i in range(1, int(math.sqrt(hcf) + 1)):
        if hcf % i == 0:
          Divisors.append(i)
          if i*i != hcf:
                Divisors.append(int(hcf / i))
  
  return Divisors

# Finds number of permutations of objects
def FindPerms(aList):
  unique = 1
  
  for element in aList:
    unique *= math.factorial(element)
 
  numPerms = math.factorial(sum(aList))

  numRealPerms = numPerms/unique

  return numRealPerms

# Asks for input
def InToList():
  numTypes   = int(input("Number of Object Types: "))
  List = []

  for i in range(numTypes):
    num = int(input("Number of Type " + str(i+1) + " Objects: "))
    List.append(num)
    
  return List

# Finds the number of permutations under rotational symmetry
def ListToNumCirc(aList):
  Helper = []
  CircularDict = {}
  
  DivisorList = FindDivisors(aList)
  DivisorList.sort(reverse=True)

  RepTypeList = [[element / DivisorList[i] for element in aList] for i in range(len(DivisorList))]
  PermRepList = [FindPerms(rep) for rep in RepTypeList]

  PermRepDict = {DivisorList[i]: PermRepList[i] for i in range(len(DivisorList))}

  for i in range(len(DivisorList)):
    Helper = FindDivisors([DivisorList[i]])
    Helper.remove(DivisorList[i])
    for divisor in Helper:
      PermRepDict[divisor] = PermRepDict[divisor] - PermRepDict[(DivisorList[i])]
      
  for key in PermRepDict:
    x = key/(sum(aList))
    CircularDict[key]= PermRepDict[key]*x
  
  return sum(list(CircularDict.values()))
  
ListToNumCirc(InToList())


```
